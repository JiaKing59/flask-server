```python
# Path where the data set is located
dataset_path = 'flower1'

# Ratio of training set to test set
test_frac = 0.2

# All categories
classes = os.listdir(dataset_path)

# Create train and test directories
os.makedirs(os.path.join(dataset_path, 'train'), exist_ok=True)
os.makedirs(os.path.join(dataset_path, 'val'), exist_ok=True)

df = pd.DataFrame()

print('{:^18} {:^18} {:^18}'.format('type', 'Number of data in the training set', 'Number of data in the test set'))

for flower in classes: # Iterate through each category

    # Retrieve the names of all image files in the category
    old_dir = os.path.join(dataset_path, flower)
    images_filename = os.listdir(old_dir)
    random.shuffle(images_filename) # Random disruptions

    # Dividing the training and test sets
    testset_numer = int(len(images_filename) * test_frac) # Number of test set images
    testset_images = images_filename[:testset_numer]      # Get the name of the test set image file to be moved to the test directory
    trainset_images = images_filename[testset_numer:]     # Get the image file name of the training set to be moved to the train directory

    # Moving images to the test directory
    for image in testset_images:
        old_img_path = os.path.join(dataset_path, flower, image)         # Get the original file path
        new_test_path = os.path.join(dataset_path, 'val', flower, image) # Gets the new file path to the test directory
        shutil.move(old_img_path, new_test_path) # Moving files

    # Moving images to the train directory
    for image in trainset_images:
        old_img_path = os.path.join(dataset_path, flower, image)           # Get the original file path
        new_train_path = os.path.join(dataset_path, 'train', flower, image) # Get the new file path to the train directory
        shutil.move(old_img_path, new_train_path) # Moving files
    
    # delete old folder
    assert len(os.listdir(old_dir)) == 0 # Ensure that all images in the old folder are moved away
    shutil.rmtree(old_dir) # Delete Folder
    
    # Outputs the number of data in each category in a neat manner
    print('{:^18} {:^18} {:^18}'.format(flower, len(trainset_images), len(testset_images)))
    
    # Save to form
    df = pd.concat([df, pd.DataFrame({'class': [flower], 'trainset': [len(trainset_images)], 'testset': [len(testset_images)]})])


# Rename dataset folder
shutil.move(dataset_path, dataset_name+'_split')

# Statistical table of the number of categories in the dataset, exported as a csv file
df['total'] = df['trainset'] + df['testset']
df.to_csv('Data volume statistics.csv', index=False)
```

```python
import time
import os

import numpy as np
from tqdm import tqdm

import torch
import torchvision
import torch.nn as nn
import torch.nn.functional as F

import matplotlib.pyplot as plt
%matplotlib inline


import warnings
warnings.filterwarnings("ignore")
```

```python
#Use GPU if available, otherwise use CPU.
device = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')
print('device', device)
```

```python
from torchvision import transforms

# Training set image pre-processing: scaling and cropping, image enhancement, conversion to Tensor, normalization
train_transform = transforms.Compose([transforms.RandomResizedCrop(224),
                                      transforms.RandomHorizontalFlip(),
                                      transforms.ToTensor(),
                                      transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
                                     ])

# Test set image pre-processing - RCTN: scaling, cropping, conversion to Tensor, normalisation
test_transform = transforms.Compose([transforms.Resize(256),
                                     transforms.CenterCrop(224),
                                     transforms.ToTensor(),
                                     transforms.Normalize(
                                         mean=[0.485, 0.456, 0.406], 
                                         std=[0.229, 0.224, 0.225])
                                    ])
```

```python
# Path to dataset folder
dataset_dir = 'flower'
```

```python
train_path = os.path.join(dataset_dir, 'train')
test_path = os.path.join(dataset_dir, 'val')
print('training set path', train_path)
print('test set path', test_path)
```

```python
from torchvision import models
import torch.optim as optim
```

```python
model = models.resnet18(pretrained=True) # Loading pre-trained models

# Modify the full join layer so that the output of the full join layer corresponds to the number of categories in the current dataset
# Default for newly created layers requires_grad=True
model.fc = nn.Linear(model.fc.in_features, n_class)
```

```python
model.fc
```

```python
# Fine tune only the parameters of the last fully connected layer of training, the other layers are frozen Adaptive Moment Estimation)
optimizer = optim.Adam(model.fc.parameters())
```

```python
model = model.to(device)

# Cross-entropy loss function
criterion = nn.CrossEntropyLoss() 

# Training rounds Epoch
EPOCHS = 20
```

```python
# Get data and annotations for a batch
images, labels = next(iter(train_loader))
images = images.to(device)
labels = labels.to(device)
```

```python
# Input model, perform forward prediction
outputs = model(images)
```

```python
# Get the predicted category logit scores for all images in the current batch
outputs.shape
```

```python
# From logit, calculate the average cross-entropy loss function value for each sample in the current batch
loss = criterion(outputs, labels)
```

```python
# Reverse communication in "three parts"
optimizer.zero_grad() # Clear gradient
loss.backward() # Reverse propagation
optimizer.step() # Optimisation updates
```

```python
# Get the predicted categories for all images in the current batch
_, preds = torch.max(outputs, 1)
```

```python
preds
```

```python
labels
```

```python
# Iterate through each EPOCH
for epoch in tqdm(range(EPOCHS)):

    model.train()

    for images, labels in train_loader:  # Get a batch of the training set, containing data and annotations
        images = images.to(device)
        labels = labels.to(device)

        outputs = model(images)           # forward prediction, get the prediction of the current batch
        loss = criterion(outputs, labels) # Compare the prediction results with the annotation and calculate the cross-entropy 
                                          # loss function for the current batch
        
        optimizer.zero_grad()
        loss.backward()                   # Loss function back-propagation of neural network weights to find the gradient
        optimizer.step()                  # Optimized update of neural network weights
```

```python
model.eval()
with torch.no_grad():
    correct = 0
    total = 0
    for images, labels in tqdm(test_loader): # Get a batch of the test set, containing data and annotations
        images = images.to(device)
        labels = labels.to(device)
        outputs = model(images)              # Forward prediction to obtain confidence in the prediction of the current batch
        _, preds = torch.max(outputs, 1)     # Obtain the category corresponding to the maximum confidence level as the prediction result
        total += labels.size(0)
        correct += (preds == labels).sum()   # Number of correct samples predicted
    print('The accuracy on the test set is {:.3f} %'.format(100 * correct / total))
```

```python
import torch
import onnx
import tensorflow as tf
import onnx_tf
from torchvision.models import resnet50

# Load the PyTorch model
pytorch_model_path = 'flower_pytorch_C1.pt'
pytorch_model = torch.load(pytorch_model_path)
pytorch_model.eval()

# Export the PyTorch model to ONNX format
input_shape = (1, 3, 224, 224)
dummy_input = torch.randn(input_shape)
onnx_model_path = 'flower_model.onnx'
torch.onnx.export(pytorch_model, dummy_input, onnx_model_path, verbose=False)

# Load the ONNX model
onnx_model = onnx.load(onnx_model_path)

# Convert the ONNX model to TensorFlow format
tf_model_path = 'flower_model.pb'
tf_rep = onnx_tf.backend.prepare(onnx_model)
tf_rep.export_graph(tf_model_path)

# Convert the TensorFlow model to TensorFlow Lite format
converter = tf.compat.v1.lite.TFLiteConverter.from_saved_model(tf_model_path)
tflite_model = converter.convert()

# Save the TensorFlow Lite model to a file
with open('flower_model.tflite', 'wb') as f:
    f.write(tflite_model)
```

```python
import sys
import os
import numpy as np
import cv2
import torch
from torchvision.transforms import transforms
from PyQt5 import QtCore, QtGui, QtWidgets


# Define class names
class_names = ['Chysanthermun flower','Lily flower','Lotus flower','Rafflesia flower','Rose flower','Tulip flower']

class MainWindow(QtWidgets.QMainWindow):
    def __init__(self):
        super().__init__()
        self.setWindowTitle("Image Classifier")
        self.setWindowIcon(QtGui.QIcon("icon.png"))
        self.setGeometry(100, 100, 600, 700)  # Change the width and height

        # Create widgets
        self.image_label = QtWidgets.QLabel(self)
        self.image_label.setGeometry(QtCore.QRect(50, 50, 500, 500))
        self.image_label.setStyleSheet("background-color: black; border: 1px solid black;")
        self.select_button = QtWidgets.QPushButton("Select Image", self)
        self.select_button.setGeometry(QtCore.QRect(200, 580, 200, 50))  # Increase the size of the button
        self.select_button.setStyleSheet("background-color: #5cb85c; color: white; font-size: 20px; font-weight: bold;")
        self.select_button.clicked.connect(self.select_image)

        # Create label for prediction result
        self.prediction_label = QtWidgets.QLabel(self)
        self.prediction_label.setGeometry(QtCore.QRect(50, 650, 500, 50))  # Center the label
        self.prediction_label.setAlignment(QtCore.Qt.AlignCenter)
        self.prediction_label.setStyleSheet("font-size: 28px; font-weight: bold;")

        # Load model
        self.model = torch.load("flower_pytorch_C1.pt")
        self.model.eval()

        # Set application style sheet
        self.setStyleSheet("QMainWindow {background-color: white; font-family: Arial, sans-serif;}")



    def select_image(self):
        # Open file dialog to select image
        file_dialog = QtWidgets.QFileDialog()
        file_dialog.setNameFilters(["Images (*.png *.xpm *.jpg *.bmp)"])
        file_dialog.selectNameFilter("Images (*.png *.xpm *.jpg *.bmp)")
        file_path = ""
        if file_dialog.exec_():
            file_path = file_dialog.selectedFiles()[0]

        if file_path:
            # Load and preprocess image
            image = cv2.imread(file_path)
            image = cv2.resize(image, (224, 224))
            image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
            image = image.astype(np.float32) / 255.0

            # Apply transforms
            transform = transforms.Compose([
                transforms.ToTensor(),
                transforms.Normalize(mean=[0.485, 0.456, 0.406],
                         std=[0.229, 0.224, 0.225])
                ])
            image = transform(image)
            image = image.unsqueeze(0)


            # Make prediction
            with torch.no_grad():
                prediction = self.model(image)
            prediction = prediction.argmax().item()
            prediction_name = class_names[prediction]

            # Display image and prediction result
            pixmap = QtGui.QPixmap(file_path)
            self.image_label.setPixmap(pixmap.scaled(650, 650, QtCore.Qt.KeepAspectRatio))
            self.prediction_label.setText(f"Prediction: {prediction_name}")


if __name__ == "__main__":
    # Create application and window
    app = QtWidgets.QApplication(sys.argv)

    # Set application style sheet
    app.setStyleSheet("QMainWindow {background-color: white;}")

    # Create window
    window = MainWindow()
    window.show()

    # Run application
    sys.exit(app.exec_())
```
